```python
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# 와인 데이터를 로드한다.
wine = load_wine()
# 와인 데이터에서 feature 로 되어있는 데이터를 갖고 온다.
wine_data = wine.data
# 와인 데이터에서 target 데이터를 갖고온다.
wine_target = wine.target
# wine 데이터의 feature, target로 Dataframe을 만든다.
df_wine = pd.DataFrame(data = wine_data, columns=[wine.feature_names])
# feature 만 있는 데이터 프레임에 타겟 데이터를 추가한다.
df_wine['target'] = wine.target
df_wine

x_training, x_test, y_training, y_test = train_test_split(wine_data, wine_target, test_size=0.3)
#train_test_split의 첫번째 데이터인 wine_data는 feature의 데이터이고
#두번째 데이터인 wine_target은 target 데이터세트이다.
#test_size는 전체 데이터 중에서 테스트 데이터의 비율이다. 현재는 0.3으로 해놨으니 0.7은 학습 데이터이고 0.3은 테스트 데이터이다.

dtc = DecisionTreeClassifier()
dtc.fit(x_training, y_training)
#DecisionTreeClassifier 객체를 형성
#생성된 DecisionTreeClassifier 객체의 fit() 을 이용해서 학습용 데이터의 feature 데이터, target 데이터를 이용하여서 학습을 수행한다.

predict = dtc.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predict))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pandas as pd

# 유방암 데이터를 로드한다.
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_feature_names = breast_cancer.feature_names
breast_cancer_label = breast_cancer.target
breast_cancer_label_names = breast_cancer.target_names

print(f'라벨 종류 : {breast_cancer_label_names}')
print(f'{breast_cancer_data.shape[0]}개의 데이터를 {len(breast_cancer_label_names)}가지로 분류하는 모델이 필요')

breast_cancer_df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer_feature_names)
breast_cancer_df

x_train, x_test, y_train, y_test = train_test_split(breast_cancer_data, breast_cancer_label, test_size=0.3, random_state=20)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

# Decision Tree

from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=20)
print(decision_tree._estimator_type)
print('DecisionTreeClassifier 분류 모델이다.')

# 모델 학습
decision_tree.fit(x_train, y_train)
# 예측하기
y_pred_decision = decision_tree.predict(x_test)
# 정확도
print('\n')
print('정답률 :', accuracy_score(y_test, y_pred_decision))
# 레포트
print('\n')
print(classification_report(y_test, y_pred_decision))

# 서포트 벡터 머신(SVM, Support Vector Machine)

import sklearn.svm as svm

svm = svm.SVC()
print(svm._estimator_type)
print('Support Vector Machine 분류 모델이다.')
svm.fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
print('\n')
print('정답률 :', accuracy_score(y_test, y_pred_svm))
print('\n')
print(classification_report(y_test, y_pred_svm))

# 확률적 경사 하강법 (SGD, Stochastic Gradient Descent)

from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()
print(sgd._estimator_type)
print('Stochastic Gradient Descent 분류 모델이다.')
sgd.fit(x_train, y_train)
y_pred_sgd = sgd.predict(x_test)
print('\n')
print('정답률 :', accuracy_score(y_test, y_pred_sgd))
print('\n')
print(classification_report(y_test, y_pred_sgd))

# Logistic Regression

from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression(max_iter=7000)
print(logistic._estimator_type)
print('LogisticRegression 분류 모델이다.')
logistic.fit(x_train, y_train)
y_pred_logistic = logistic.predict(x_test)
from sklearn.metrics import accuracy_score
print('\n')
print('정답률 :', accuracy_score(y_test, y_pred_logistic))
print('\n')
print(classification_report(y_test, y_pred_logistic))
```
